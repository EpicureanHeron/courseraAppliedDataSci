```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# fb_df.head()
# https://en.wikipedia.org/wiki/List_of_Michigan_Wolverines_football_seasons
# https://en.wikipedia.org/wiki/List_of_Michigan_Wolverines_men%27s_ice_hockey_seasons
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def hockey_data():
    # read the excel file and drop head and footer
    h_df = pd.read_excel('hockey_seasons.xlsx',header=2, skipfooter=7)
#     list of columns to drop
    columns_to_drop = ['Unnamed: 1', 'GP', "W", "L", "T", "3/SW", "Pts*", 'Finish', 'GP.1', 'W.1', 'L.1', 'T.1' , 'Unnamed: 14', 'Unnamed: 15']
#     drops the aforementioned columns
    h_df_years = h_df.drop(columns_to_drop, axis=1)
# drop any leftover fields with NaN in it
    h_df_years = h_df_years.dropna()
#     the season spans two calendar years, but for purposes of merging with football data, only one
#  the first year as the index, so we use lambda to do that 
    h_df_years['Year'] = h_df_years['Unnamed: 0'].apply(lambda x: str(x)[0:4])
#     set the index to be new year column
    h_df_years['Year'] = pd.to_numeric(h_df_years['Year'])
    h_df_years = h_df_years.set_index('Year')
#     drops the old season spanning years
    h_df_years = h_df_years.drop('Unnamed: 0', axis=1)
#     renames the % to winning percent
    h_df_years.rename(columns = {'%':'Hockey Win %'}, inplace = True)
    # returns the data frame
    return h_df_years
    
def football_data():
    
    fb_df = pd.read_excel('football_seasons.xlsx')   
    columns = ['Coach', 'Conference', 'Standing', 'Bowl/playoffs',
       'Coaches#', 'AP°', 'Unnamed: 8', 'Unnamed: 9']   
    fb_df = fb_df.drop(columns, axis=1)   
    fb_df = fb_df.dropna()  
    fb_df['Football %'] = ''
    fb_df = fb_df.reset_index()
    fb_df = fb_df.drop('index', axis=1)
    record_list = fb_df['Overall'].tolist()
    fb_df= fb_df.drop(fb_df.index[141])
  
    index_value = 0
    for record in record_list:
        record_values_list = str(record).split('–')
#        list compression for the win, makes everything within the list an integer
        record_values_list = [int(i) for i in record_values_list]
        total_games = sum(record_values_list)
        games_won = record_values_list[0]
        if total_games > 0:
            percentage_won = games_won/total_games

        else:
            
            percentage_won = 0
            
        fb_df.at[index_value, 'Football %'] = percentage_won
        
        index_value += 1
#     https://stackoverflow.com/questions/37600711/pandas-split-column-into-multiple-columns-by-comma
    fb_df['Year'] = pd.to_numeric(fb_df['Year'])
    fb_df = fb_df.set_index('Year')
    
    fb_df = fb_df.drop('Overall', axis= 1)
    return fb_df

f = football_data()

h = hockey_data()

hf = pd.merge(h, f, left_index=True, right_index=True, how='inner')

hf = hf.reset_index()

hf = hf[hf['Year'] >= 1980]

hf = hf.set_index('Year')

def plot_data (df):
    
    plt.figure()
    
    plt.plot(df)
    
    ax = plt.gca()
    fig = plt.gca()
    ax.legend(['Hockey','Football'])
    plt.ylabel('Win %')
    plt.xlabel('Year')
    plt.title("University of Michigan Men's Sports Win Ratio Since 1980")
    line_y = .5
    hoz_line = plt.axhline(y=line_y, color='grey', linewidth=2, linestyle='--');
    y_text = plt.text(2020.5, line_y, '50% Thershold', bbox=dict(fc='white',ec='k'));

    return plt

plot_data(hf)

```
